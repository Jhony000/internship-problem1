```python
import zipfile
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Extract ZIP file
zip_path = "Downloads/API_SP.POP.TOTL_DS2_en_csv_v2_73852.zip"
extract_path = "population_data"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Step 2: List extracted files (to see available data files)
extracted_files = os.listdir(extract_path)
print("Extracted files:", extracted_files)

# Step 3: Load the main CSV file, skipping metadata rows
csv_file_path = os.path.join(extract_path, 'API_SP.POP.TOTL_DS2_en_csv_v2_73852.csv')
df_population = pd.read_csv(csv_file_path, skiprows=4)

# Step 4: Preview the DataFrame
print(df_population.head())

# Step 5: Optional - Plot histogram (if you have age data; else this is placeholder)
# Simulate data if "age" column not present
import numpy as np

np.random.seed(42)
ages = np.random.normal(loc=35, scale=10, size=200).astype(int)

# Plot histogram
plt.figure(figsize=(10, 5))
sns.histplot(ages, bins=15, kde=True, color='skyblue')
plt.title('Age Distribution in Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Simulate gender data
np.random.seed(1)
genders = np.random.choice(['Male', 'Female', 'Other'], size=200, p=[0.45, 0.45, 0.10])

# Create DataFrame
df = pd.DataFrame({'Gender': genders})

# Plot bar chart
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title('Gender Distribution in Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
# Simulate age data (normally distributed around 35 years)
ages = np.random.normal(loc=35, scale=10, size=200).astype(int)

# Create DataFrame
df_age = pd.DataFrame({'Age': ages})

# Plot histogram
plt.figure(figsize=(8, 4))
sns.histplot(df_age['Age'], bins=15, kde=True, color='skyblue')
plt.title('Age Distribution in Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
